## Constructor

**Constructor is a special type of method which allows the user to give the parameters at the time of object creation**

```python
class Employee:
    def __init__(self,name,age,salary,gender): # __init__ method act as a constructor
        self.name = name
        self.age = age
        self.salary = salary
        self.gender = gender
    def employee_details(self):
        print(f'Name : {self.name}')
        print(f'Age : {self.age}')
        print(f'Salary : {self.salary}')
        print(f'Gender : {self.gender}')
```

```python
ak = Employee('Akash',22,'5.2L','Male')
ja = Employee('Jay',20,'5.2L','Male')
```

```python
ak.employee_details()
```

```python
ja.employee_details()
```

# Inheritance In Python :

**With inheritance ,one class can derive the properties of another class**
> One class is known as 'Parent Class/Base Class'
> and Derived Class is known as 'Child Class'**

#### Simple Inheritance :
Inheritance allows us to define a class that inherits all the methods and properties from another class. Parent class is the class being inherited from, also called base class. Child class is the class that inherits from another class, also called derived class.

```python
# Parent Class or Base Class
class Vehicle:
    def __init__(self,mileage,cost):
        self.mileage = mileage
        self.cost= cost
    def show_details(self):
        print('I am a Vehicle')
        print(f'Mileage : {self.mileage}')
        print(f'Cost : {self.cost}')
```

```python
v1 = Vehicle(85,'Rs.50,000')
v1.show_details()
```

```python
# Derived or Child Class
class Car(Vehicle):
    def show_car(self):
        print('I am a car')
```

```python
c1 = Car(500,'Rs.8,50,000')
c1.show_details()
```

```python
c1.show_car()
```

# Over-loading init method(constructor)

> Over-loading means that two or more methods have same 'Name' and 'Different Parameters'.

```python
# Parent Class or Base Class
class Vehicle:
    def __init__(self,mileage,cost):
        self.mileage = mileage
        self.cost= cost
    def show_details(self):
        print('I am a Vehicle')
        print(f'Mileage : {self.mileage}')
        print(f'Cost : {self.cost}')
```

```python
v1 = Vehicle(85,'Rs.50,000')
```

```python
class Car(Vehicle):
    def __init__(self,mileage,cost,tyres,hp): # over loading __init__ method
        super().__init__(mileage,cost)
        self.tyres = tyres
        self.hp= hp
    def show_car_details(self):
        print('I am a car')
        print(f'No of tyres : {self.tyres}')
        print(f'Horse Power : {self.hp}')
```

```python
c1= Car(300,'Rs.7.0L',4,200)
c1.show_details()
```

```python
c1.show_car_details()
```

#### Multiple Inheritance:
When a class is derived from more than one base class it is called multiple Inheritance. 

```python
# Parent Class One
class Parent1():
    def assign_string_one(self,str1):
        self.str1 = str1
    def show_string_one(self):
        return self.str1
```

```python
# Parent Class Two
class Parent2():
    def assign_string_two(self,str2):
        self.str2 = str2
    def show_string_two(self):
        return self.str2
```

```python
# Child Class
class Derived(Parent1,Parent2):
    def assign_string_three(self,str3):
        self.str3 = str3
    def show_string_three(self):
        return self.str3
```

```python
ch = Derived()
ch.assign_string_one('I am parent 1')
ch.show_string_one()
```

```python
ch.assign_string_two('I am parent 2')
ch.show_string_two()
```

```python
ch.assign_string_three('I am child')
ch.show_string_three()
```

#### Multi-level Inheritance
In the Multilevel inheritance, a derived class will inherit a base class and as well as the derived class also act as the base class to other class.

```python
# Parent class
class Parent():
    def assignName(self,name):
        self.name = name
    def showName(self):
        return self.name
```

```python
# Child Class
class Child(Parent):
    def assignAge(self,age):
        self.age = age
    def showAge(self):
        return self.age
```

```python
# Grand Child
class GrandChild(Child):
    def assignGender(self,gender):
        self.gender = gender
    def showGender(self):
        return self.gender
```

```python
gch = GrandChild()
gch.assignName('Akash')
gch.assignAge(22)
gch.assignGender('Male')
```

```python
gch.showName()
```

```python
gch.showAge()
```

```python
gch.showGender()
```

#### Hybrid Inheritance
Hybrid Inheritance is a blend of more than one type of inheritance.
